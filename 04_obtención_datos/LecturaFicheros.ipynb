## Ficheros delimitados

Frecuentemente nos encontraremos con ficheros que contienen muchos datos por cada línea. Estos ficheros suelen estar separados por _comas_ o _tabs_. Cada línea puede contener varios campos, los cuales utilizarán un separador indicando cuando finaliza y comienza un nuevo campo. Asimismo, hay que lidiar con saltos de línea.

**CSV = comma separated value**

CSV es un estándard para almacenar datos en formato de texto, donde las comas se utilizan para separar distintas columnas y el salto de linea se utiliza para separar filas. Normalmente, la primera fila de un csv contiene los nombres de columnas para los datos. Además de las comas, el tabulador (\t) y el ";" son los carácteres que se utilizan para delimitar los campos.


Para evitar complicaciones y fallos en la lectura de ficheros, se puede hacer uso del modulo _*csv*_ de Python o de la librería _pandas_

Si ejecutáis ese código, nos devolverá un error. Necesitamos abrir el csv en modo texto:

```python
import csv

def processData(name,age, wage):
    print("%s is %s and earns %s per week"%(name, age, wage))  

with open("data/data.csv", "r", encoding="utf8") as f:
    reader = csv.reader(f, delimiter=',')
    next(reader, None)  # nos saltamos la cabecera
    for row in reader:
        name = row[2]       
        age = row[3]       
        wage = row[12]       
        processData(name,age,wage)
```

En vez de acceder las columnas por índice, podríamos accederlas mediante un diccionario:

```python
import csv

def processData(name,age, wage):
    print("%s is %s and earns %s per week"%(name, age, wage))  

with open("data/data.csv", "r", encoding="utf8") as f:
    reader = csv.DictReader(f, delimiter=',')   
    for row in reader:
        name = row["Name"]       
        age = row["Age"]       
        wage = row["Wage"]       
        processData(name,age,wage)
```

Si trabajamos con la librería Pandas, podríamos utilizar el siguiente código:

```python
import pandas as pd

df = pd.read_csv('data/data.csv',sep = ",", quotechar = '"')
for index, row in df.iterrows():
    processData(row['Name'], row['Age'], row['Wage'])
```

## Otros métodos para la obtención de datos

Existen otros métodos que permiten obtener datos de otras fuentes de datos, siendo uno de los casos más interesantes, las páginas web. Entre otras técnicas, podemos utilizar el **scrapeo de webs** (lo veréis en profundidad en la asignatura Inteligencia de Datos) o la **utilización de APIs**

Ejemplo de uso de API (sin autenticación):

```python
import requests, json

endpoint = "https://api.github.com/users/moar/repos"

repos = json.loads(requests.get(endpoint).text)

print(repos)
```
