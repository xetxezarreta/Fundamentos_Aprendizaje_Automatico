#### Correlation entre atributos


Vamos a ver qué variables están correlacionadas con `SalePrice` (correlación > 0.5), que sería la variable a predecir en un hipotético modelo. Guardaremos dichas variables en una lista llamada `golden_features_list`. Reutilizaremos el dataset `df_num` para hacerlo

```python
df_num_corr = df_num.corr()['SalePrice'][:-1] # -1 because the latest row is SalePrice
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with SalePrice:\n{}".format(len(golden_features_list), golden_features_list))
```

Tenemos una lista de valores fuertemente correlacionados. Pero esta lista no es perfecta, ya que la correlación podría estar afectada por los outliers que hemos citado anteriormente. Podríamos proceder de la sigueinte manera:

 - Graficamos las variables numéricas y vemos cuáles pueden tener algunos outliers
 - Borrar los outliers de estas variables y comprobar si siguen teniendo correlación con la variable a predecir

Recordad que el factor de correlación por sí sola no siempre explica la relación entre variables, pero graficarlas puede ayudar a descubrir tendencias.

Por ejemplo, una relación curvilínea no se podría descubrir viendo solo el factor de correlación.

```python
for i in range(0, len(df_num.columns), 5):
    sns.pairplot(data=df_num,
                x_vars=df_num.columns[i:i+5],
                y_vars=['SalePrice'])
```

Estas gráficas ya nos dejan ver varias relaciones. La mayoría de variables tienen una relación linear con *SalePrice*. Si miramos más atentamente a los datos, existen varias gráficas donde hay varios datos en el x= 0, lo cual puede indicar que existen varias casas que no contengan esa característica en concreto.

Si tomamos como ejemplo `OpenPorchSF`, es dudoso que todas las casas tengan un porche.

Vamos a remover esos valores `0` y repitamos el proceso de buscar valores correlacionados:

```python
import operator

individual_features_df = []
for i in range(0, len(df_num.columns) - 1): # -1  porque la última columna es SalePrice
    tmpDf = df_num[[df_num.columns[i], 'SalePrice']]
    tmpDf = tmpDf[tmpDf[df_num.columns[i]] != 0]
    individual_features_df.append(tmpDf)

all_correlations = {feature.columns[0]: feature.corr()['SalePrice'][0] for feature in individual_features_df}
all_correlations = sorted(all_correlations.items(), key=operator.itemgetter(1))
for (key, value) in all_correlations:
    print("{:>15}: {:>15}".format(key, value))
```

Hemos encontrado otra correlación con esta pequeña limpieza... Ahora `golden_features_list` tendría esta pinta:

```python
golden_features_list = [key for key, value in all_correlations if abs(value) >= 0.5]
print("Hay {} variables fuertemente correlacionadas con SalePrice:\n{}".format(len(golden_features_list), golden_features_list))
```

## Análisis multiraviable

En este caso en vez de graficas todas las parejas de variables, graficaremos todo en una sola gráfica, ya que puede ser más fácil de interpretar.

```python
corr = df_num.drop('SalePrice', axis=1).corr() # We already examined SalePrice correlations
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

Muchas características parecen estar correlacionadas entre ellas, pero parece que `YearBuild`/`GarageYrBlt` estén correlacionadas dada la inflación de los últimos años.Para `1stFlrSF`/`TotalBsmtSF`, es normal que el primer piso sea más grande.

Para otras relaciones menos obvias:
- Hay una correlación negativa `BsmtUnfSF` (Unfinished square feet of basement area) y `BsmtFinSF2` (Type 2 finished square feet). Hay una definición para "unfinished square feet" [aquí](http://www.homeadvisor.com/r/calculating-square-footage/).
- `HalfBath`/`2ndFlrSF` esta relación es interesante y puede indicar que la gente da importancia a no tener que bajar escaleras para ir al cuarto de baño.

Seguro que existen muchas más relaciones, solo se han descrito las más obvias y se podría exprimir el dataset mucho más. De todas formas, los algoritmos de ML son mucho más efectivos a la hora de encontrar relaciones "ocultas" entre variables.

## Relación entre variables cualitativas

Podemos examinar las variables cuantitativas y ver sus relaciones con  `SalePrice`, la cual es también cuantitativa.

```python
quantitative_features_list = ['LotFrontage', 'LotArea', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'TotalBsmtSF', '1stFlrSF',
    '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
    'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 
    'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'SalePrice']
df_quantitative_values = df[quantitative_features_list]
df_quantitative_values.head()
```

Como todavía hay muchas variables por analizar, cogeremos solo las que están altamente correlacionadas y las analizaremos de uno en uno:

```python
features_to_analyse = [x for x in quantitative_features_list if x in golden_features_list]
features_to_analyse.append('SalePrice')
features_to_analyse
```

Las variables `TotalBsmtSF`, `1stFlrSF`, `GrLivArea` tienen una gran dispersión, aunque existe correlación, no parece que sea fuerte.
