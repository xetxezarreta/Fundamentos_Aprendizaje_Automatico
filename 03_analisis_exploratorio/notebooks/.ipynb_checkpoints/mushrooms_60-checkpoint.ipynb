## Análisis descriptivo

```python
sns.set(style="darkgrid")
fig,axs=plt.subplots(nrows=8,ncols=3,figsize=(30, 75))

i=0
j=0
k=0

for col in data.columns:
    
    i=int(k/3)
    j=k%3
    
    axe=sns.countplot(x=col, hue="class", data=data,ax=axs[i][j]) # for Seaborn version 0.7 and more
    
    bars = axe.patches
    half = int(len(bars)/2)
    left_bars = bars[:half]
    right_bars = bars[half:]

    for left, right in zip(left_bars, right_bars):
        height_l = np.nan_to_num(left.get_height())
        height_r = np.nan_to_num(right.get_height())
        total = height_l + height_r

        axe.text(left.get_x() + left.get_width()/2., height_l + 40, '{0:.0%}'.format(height_l/total), ha="center")
        axe.text(right.get_x() + right.get_width()/2., height_r + 40, '{0:.0%}'.format(height_r/total), ha="center")
    
    k=k+1
```

```python
# Analizando las variables podemos estimar que las variables:
# 'bruises', 'odor', 'gill-size', 'gill-color', 'spore-print-color'

# Pueden ser las variable mas significativas ya que:
# 1. El data set contiene en la variable bruises (bruises=t), una probabilidad de setas comestibles del 82%. (3376 filas son de tipo t)
# 2. La variable odor(odor=n), la probabilidad de ser comestible es del 97% ( 3528 filas son de tipo n).
# 3. En la variable gill-size(gill-size=n), la probabilidad de ser venenosa es del 89%.
# 4. La variable gill-color (gill-color=b), la probabilidad de ser venenosa es del 100% (1728 filas son de tipo b)
# 5. La variable spore-print-color, as variables k y n tiene un porcentaje altos de ser comestibles y en sus variables h y w de ser venenosas
```
